```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#school_data.head(5)
#student_data.head(5)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(5)
#school_data_complete.dtypes
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
#schools = school_data_complete["school_name"].unique()  -> shows the array with the list of unique schools
schools = school_data_complete["school_name"].nunique()
schools

#Calculate the total number of students
students = school_data_complete["Student ID"].nunique()
students

#Calculate the total number of budgets per school
schoolBudget = school_data_complete["budget"].unique() # -> shows the array with the list of unique BUDGETS

# add those values
ttlbudget = schoolBudget.sum()
print(ttlbudget)

#Calculate the average math score
avgMath = school_data_complete["math_score"].mean()
avgMath

#Calculate the average reading score
avgRead = school_data_complete["reading_score"].mean()
avgRead

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
ovPassRate = (avgMath + avgRead)/2
ovPassRate

#Calculate the percentage of students with a passing math score (70 or greater)
math70 = school_data_complete.loc[school_data_complete["math_score"] >= 70] #.count 
#math70
math70Count = len(math70)
math70Count

# Math Percentage 
mathPct = (math70Count/students)*100
mathPct

#Calculate the percentage of students with a passing reading score (70 or greater)

read70 = school_data_complete.loc[school_data_complete["reading_score"] >= 70] 
#read70
read70Count = len(read70)
read70Count

# Read Percentage 
readPct = (read70Count/students)*100
readPct

#Create a dataframe to hold the above results
summary = [{"Total Schools": schools, "Total Students": students, "Total Budget": ttlbudget, 
           "Average Math Score": avgMath, "Average Reading Score": avgRead, "%Passing Math": mathPct, 
           "%Passing Reading": readPct, "%Overall Passing Rate": ovPassRate}]
#summary

DistricSummarydf = pd.DataFrame(summary)
#DistricSummarydf

#Optional: give the displayed data cleaner formatting
format_mapping={"Total Students": '{:,.0f}', "Total Budget": '${:,.2f}', 
                "Average Math Score": '{:.6f}',
                "Average Reading Score": '{:.6f}',
                "%Passing Math": '{:.6f}',
                "%Passing Reading": '{:.6f}',
                "%Overall Passing Rate": '{:.6f}'}

for key, value in format_mapping.items():
    DistricSummarydf[key] = DistricSummarydf[key].apply(value.format)


DistricSummarydf
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Create an overview table that summarizes key metrics about each school, including:

# GroupBy School

# Using GroupBy in order to separate the data into fields according to "school" values
eachSchool = school_data_complete.groupby(['school_name'])

# In order to be visualized, a data function must be used...
eachSchool.count().head(15)

#School Name 
eachSchoolName = eachSchool["school_name"].unique()
eachSchoolName
#eachSchoolName.head(15)

#School Type
eachSchoolType = eachSchool["type"].unique()
eachSchoolType
#eachSchoolType.head(15)

#Total Students
eachSchoolStudents = eachSchool["Student ID"].count()
eachSchoolStudents.head()

#Total School Budget
eachSchoolBudget = eachSchool["budget"].unique()
eachSchoolBudget
#eachSchoolType.head(15)

#Per Student Budget
eachSchoolStdBgt = eachSchoolBudget/eachSchoolStudents
eachSchoolStdBgt

#Average Math Score
eachSchoolAveMath = eachSchool["math_score"].mean()
eachSchoolAveMath.head()

#Average Reading Score
eachSchoolAveRead = eachSchool["reading_score"].mean()
eachSchoolAveRead.head()


#% Passing Math
# GroupBy School & math > 70
#  Using GroupBy in order to separate the data into fields according to "school" values
eachSchoolMathPass = math70.groupby(['school_name'])
# In order to be visualized, a data function must be used...
eachSchoolMathPass.head()
# Count how many scores have occured within each grade score
eachSchoolMathPassCount = eachSchoolMathPass["math_score"].count()
eachSchoolMathPassCount
# Calculate percentage
eachSchoolMathPct = (eachSchoolMathPassCount/eachSchoolStudents)*100
eachSchoolMathPct

#% Passing Reading
# GroupBy School & read > 70
#  Using GroupBy in order to separate the data into fields according to "school" values
eachSchoolReadPass = read70.groupby(['school_name'])

# In order to be visualized, a data function must be used...
eachSchoolReadPass.head()

# Count how many scores have occured within each grade score
eachSchoolReadPassCount = eachSchoolReadPass["reading_score"].count()
eachSchoolReadPassCount

# Calculate percentage
eachSchoolReadPct = (eachSchoolReadPassCount/eachSchoolStudents)*100
eachSchoolReadPct


#Overall Passing Rate (Average of the above two)
esOvRate =(eachSchoolMathPct+eachSchoolReadPct)/2
esOvRate

#Create a dataframe to hold the above results

SchoolSummarydf = pd.DataFrame({"School Type": eachSchoolType, "Total Students": eachSchoolStudents, 
             "Total School Budget": eachSchoolBudget, "Per Student Budget": eachSchoolStdBgt, 
             "Average Math Score": eachSchoolAveMath, "Average Reading Score": eachSchoolAveRead, 
             "%Passing Math": eachSchoolMathPct, "%Passing Reading": eachSchoolReadPct, 
             "%Overall Passing Rate": esOvRate})

SchoolSummarydf
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.

# Filter the data so that only those grades in the 9th are in a DataFrame
Grade9df = school_data_complete.loc[school_data_complete["grade"]== "9th",: ]
Grade9bySchool = Grade9df.groupby(["school_name"])
#Calculate the average math score
avgMathxS9 = Grade9bySchool["math_score"].mean()
avgMathxS9

Grade10df = school_data_complete.loc[school_data_complete["grade"]== "10th",: ]
Grade10bySchool = Grade10df.groupby(["school_name"])
avgMathxS10 = Grade10bySchool["math_score"].mean()

Grade11df = school_data_complete.loc[school_data_complete["grade"]== "11th",: ]
Grade11bySchool = Grade11df.groupby(["school_name"])
avgMathxS11 = Grade11bySchool["math_score"].mean()

Grade12df = school_data_complete.loc[school_data_complete["grade"]== "12th",: ]
Grade12bySchool = Grade12df.groupby(["school_name"])
avgMathxS12 = Grade12bySchool["math_score"].mean()

#Combine the series into a dataframe
Gradesdf1 = pd.merge(avgMathxS9, avgMathxS10, on="school_name",how="left")
Gradesdf1 = Gradesdf1.rename(columns={"math_score_x": "9th","math_score_y": "10th"})
Gradesdf2 = pd.merge(avgMathxS11, avgMathxS12, on="school_name",how="left")
Gradesdf2 = Gradesdf2.rename(columns={"math_score_x": "11th","math_score_y": "12th"})
Gradesdf = pd.merge(Gradesdf1, Gradesdf2, on="school_name",how="left")
Gradesdf
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.

#Calculate the average reading score
avgReadxS9 = Grade9bySchool["reading_score"].mean()
avgReadxS10 = Grade10bySchool["reading_score"].mean()
avgReadxS11 = Grade11bySchool["reading_score"].mean()
avgReadxS12 = Grade12bySchool["reading_score"].mean()

#Combine the series into a dataframe
Gradesdf1r = pd.merge(avgReadxS9, avgReadxS10, on="school_name",how="left")
Gradesdf1r = Gradesdf1r.rename(columns={"reading_score_x": "9th","reading_score_y": "10th"})
Gradesdf2r = pd.merge(avgReadxS11, avgReadxS12, on="school_name",how="left")
Gradesdf2r = Gradesdf2r.rename(columns={"reading_score_x": "11th","reading_score_y": "12th"})
Gradesdfr = pd.merge(Gradesdf1r, Gradesdf2r, on="school_name",how="left")
Gradesdfr
```
